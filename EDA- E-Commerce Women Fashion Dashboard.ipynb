## <u style="color:green"> Add new column "Discount Range" in the dataframe</u>

### Create create_range function which is reusable

```python
def create_range(data,interval):
    data_list=list(range(0,int(data.max()),interval))
    data_range=[]
    i=0
    l=len(data_list)-1
    while (i<l):
        if(i==0):
            data_range.append(f"1-{data_list[i+1]}")
        elif(i==(l-1)):
            data_range.append(f"{data_list[i]+1}-above") 
        else:
            data_range.append(f"{data_list[i]+1}-{data_list[i+1]}")
        i=i+1
        #data_list[0]=-1
    
    return data_list,data_range
```

### add new column in the dataframe with the help of pd.cut and create_range function

```python
data=create_range(df.Discount,5)
df["Discount Range"]=pd.cut(x=df.Discount,bins=data[0],labels=data[1])
```

```python
def price_range(data):
    list_range=[]
    if(data<=1000):
        return "1000 below"
    elif(data<=5000):
        return "1001-5000"
    elif(data<=10000):
        return "5001-10000"
    elif(data<=20000):
        return "20001-50000"
    elif(data<=50000):
        return "50001-100000"
    else:
        return "100000- above"
df["Price Range"]=df.SellPrice.apply(price_range)
df.head()
```

## Which top 10 brannames, discounts and selling price ranges are available for the user?

```python
# function which going to bild a barplot of top 10 atributes, that it will receive 
def count_plots(column, topn=10):
    plt.figure()
    a=df[column].value_counts().nlargest(topn).plot(kind='bar')
    plt.title(f"{column} value counts")
    plt.ylabel('counts')
    a.set_facecolor("lightyellow")
    for count in a.containers:
        a.bar_label(count)

# list of atributes, that I'll pass to the function
columns = ['BrandName', 'Price Range','Discount']

for column in columns:
    count_plots(column)
```

### Conclusion :<br/> <p style="color:green"> <u style="color:blue">Vastranand brand</u> is most liked by the customer. The customer is more purchsed <u style="color:blue">1001-5000</u> range's product. <u style="color:blue">50% and without discount</u> are best selling products </p>

## How are the selling price, MRP price and discount distribution looks like for categories and top 10 brands

```python
#create function for data visualization

def top_visual(data,numeric):
    plt.figure(figsize=(13,4))
    #find to 10 brand
    top_brand=df[data].value_counts().nlargest(10).index.to_list()

    #fetch data of the top 10 brands
    #top_brand_df=df.loc[df["BrandName"].isin(top_brand)]
    top_brand_df=df[df[data].isin(top_brand)]

    # data visualization
    sea.boxenplot(data=top_brand_df, x=data, y=numeric).set_facecolor("lightyellow")    
```

```python
price_type=["MRP","SellPrice","Discount"]
for type in price_type:
    top_visual("BrandName",type)
```
